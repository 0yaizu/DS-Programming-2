```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_fontja
import seaborn as sns
import scipy.stats as stats
import sqlite3
import random
```

```python
# DBからimport -> DataFrame
db_file = "./rent_prices_many.db"

conn = sqlite3.connect(db_file)

df = pd.read_sql_query("SELECT * FROM rent_prices", conn)

print(df.head())
conn.close()
```

```python
# 欠損値確認
df.isnull().sum()
```

```python
# 区のリストを取得
city_list = df["city"].unique()
print(city_list)
```

```python
# 散布図を表示
sns.pairplot(df, hue="city", diag_kind="hist", size=5)
# sns.pairplot(df, hue="city", size=5)
```

```python
sns.catplot(data=df, x="city", y="rent_price", kind="bar")
```

```python
sns.boxplot(data=df, y="city", x="rent_price", whis=(0, 100))
```

```python
sns.displot(df, x="rent_price", col="city", col_wrap=3, height=5, aspect=1)
```
